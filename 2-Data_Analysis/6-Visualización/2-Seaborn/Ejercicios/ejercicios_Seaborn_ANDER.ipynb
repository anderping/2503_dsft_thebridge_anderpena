```python
# 11. Mira cuntos 'Type 1' diferentes hay en el DataFrame

pokemon['Type 1'].nunique()
```

```python
# 12. Qudate del Dataframe solo con los 'Type 1' iguales a 'Grass', 'Fire' o 'Water'

pokemon_filtered = pokemon[pokemon['Type 1'].isin(['Grass', 'Fire', 'Water'])]
pokemon_filtered
```

```python
# 13. Usa un estilo 'whitegrid'

plt.style.use('seaborn-v0_8-whitegrid')
```

```python
# 14. Pinta diagramas de violn con el DataFrame filtrado por los 3 tipos

sns.violinplot(data=pokemon_filtered);
```

```python
# 15. Escribe una lista de colores con valores '#78C850', '#F08030',  '#6890F0'

colors = ['#78C850', '#F08030', '#6890F0']
```

```python
# 16. Pinta de nuevo los diagramas de violn aadiendo esos colores en la paleta (palette)

sns.violinplot(data=pokemon_filtered, palette=colors);
```

```python
# 17. Con el diagrama de violn vemos la forma de la distribucin, pero no sabemos si Grass tiene muchos puntos y Fire tiene muy pocos.
# Para ver esto, usamos el swarmplot.
# Pinta el ltimo diagrama de violines pero con un swarmplot en lugar de un violin

sns.swarmplot(data=pokemon_filtered, palette=colors);
```

```python
# 18. Queremos ver los violines a la vez que los swarm, los podemos pintar juntos. 
# Elimina los diagramas de caja dentro de los violines con el parmetro inner = None y pon una transparencia de 0.9 en el swarmplot
# Pon un ttulo con pyplot (como siempre)

sns.violinplot(data=pokemon_filtered, palette=colors, inner=None)
sns.swarmplot(data=pokemon_filtered, alpha=0.9);
```

```python
# 19. Veamos la funcin melt() de pandas. Melt toma los valores de varias columnas y los junta todos en una misma columna nueva.
# Dado un DataFrame, melt recibe varios parmetros:
# * qu columnas del DataFrame se quedan igual
# * qu nombres de columnas del dataframe pasan a ser una columna nueva
# * cmo queremos llamar esas nuevas columnas

# Primero recuerda el Dataframe df_grass_fire_water con las primeras 5 filas

pokemon_filtered.head()
```

```python
# 20. Queremos dejar igual las variables 'Name' y 'Type 1'
# Queremos aadir una columa nueva cuyos valores sean las columnas que queremos apilar y que se llame 'New Column'
# Escribimos df_gfw.melt(id_vars=['Name', 'Type 1'], value_vars=[...], var_name='New Column')

df_grass_fire_water_melted = pokemon_filtered.melt(id_vars=['Name', 'Type 1'], value_vars=['Attack', 'Defense', 'Speed'], var_name='New Column', value_name='Value')
df_grass_fire_water_melted
```

```python
# 21. Pintamos el swarmplot con el melted y entendemos la informacin

sns.swarmplot(data=df_grass_fire_water_melted, y='Value', x='New Column', hue='Type 1', dodge=True);
```

```python
# 22. Basndote en el grfico anterior, define el tamao de la figura en (10,6) (usa plt)
# En el swarmplot aade el argumento dodge a True y usa la anterior paleta de colores Pokemon

# Dodge en ingls es "esquivar", qu diferencia ves?

# Usando plt 
# fija el lmite del eje Y entre 0 y 200
# fija la leyenda con bbox_to_anchor en upper left. Pista: upper left es el cdigo 2 y se usa el parmetro loc.

plt.figure(figsize=(10, 6))
sns.swarmplot(data=df_grass_fire_water_melted, y='Value', x='New Column', hue='Type 1', palette=colors, dodge=True)
plt.ylim(0, 200)
plt.legend(bbox_to_anchor=(1, 1), loc=2)
plt.show();
```

Al poner dodge se separan las categoras de hue.

```python
# Observa que incluso usando Seaborn, usamos Pyplot para trabajar con los ejes, la leyenda y el tamao
# Lo importante es que Seaborn y Pyplot conviven bien juntos
```
