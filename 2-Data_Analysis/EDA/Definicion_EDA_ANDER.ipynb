```python
import pandas as pd
import kagglehub

# guarda en variables los datasets y su fuente

fuente1 = "https://www.kaggle.com/datasets/tomigelo/spotify-audio-features"
tracks1_path = kagglehub.dataset_download("tomigelo/spotify-audio-features")
tracks1 = pd.read_csv(fr"{tracks1_path}\SpotifyAudioFeaturesApril2019.csv")

fuente2 = "https://www.kaggle.com/datasets/maharshipandya/-spotify-tracks-dataset"
tracks2_path = kagglehub.dataset_download("maharshipandya/-spotify-tracks-dataset")
tracks2 = pd.read_csv(fr"{tracks2_path}\dataset.csv")

fuente3 = "https://www.kaggle.com/datasets/rodolfofigueroa/spotify-12m-songs/discussion/532949"
tracks3_no_popularity_path = kagglehub.dataset_download("rodolfofigueroa/spotify-12m-songs")
tracks3_no_popularity = pd.read_csv(fr"{tracks3_no_popularity_path}\tracks_features.csv")

fuente4 = "Spotify API"
```

```python
merged_tracks = pd.merge(left=tracks1, right=tracks2, how='inner', on='track_id')
print(f"Número de filas: {len(merged_tracks)}\n")
print(f"Número de canciones no coincidentes: {len(tracks2) - len(merged_tracks)}")
merged_tracks.head()
```

```python
# Dataset completo juntando ambos:

tracks2_filtrado = tracks2[~tracks2['track_id'].isin(tracks1['track_id'])]

tracks_total = pd.concat([tracks2_filtrado, tracks1], ignore_index=True)
tracks_total
```
